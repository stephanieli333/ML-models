```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import OneHotEncoder
# from scipy.stats import spearmanr

adult = pd.read_csv('/Users/chrisrafiaa/Desktop/McGill/Winter 2020/COMP 551/comp551a1/adult_withoutEncoding.csv')
adult.head()
```

```python
adult = adult.drop(columns='Unnamed: 0')
adult = adult.drop(columns='fnlwgt')
adult.head()
```

```python
cat_features = adult.select_dtypes(include = ['category'])
fig, ax =plt.subplots(3,4)
fig.set_figheight(40)
fig.set_figwidth(40)
ax[0][0].scatter(adult['age'],adult['income'])

sns.countplot(y = 'workclass', hue = 'income', data = cat_features,ax=ax[0][1])
ax[0][2].scatter(adult['educational-num'],adult['income'])
sns.countplot(y = 'marital-status', hue = 'income', data = cat_features,ax=ax[0][3])
sns.countplot(y = 'occupation', hue = 'income', data = cat_features,ax=ax[1][0])
sns.countplot(y = 'relationship', hue = 'income', data = cat_features,ax=ax[1][1])
sns.countplot(y = 'race', hue = 'income', data = cat_features,ax=ax[1][2])
sns.countplot(y = 'gender', hue = 'income', data = cat_features,ax=ax[1][3])
ax[2][0].scatter(adult['capital-gain'],adult['income'])
ax[2][1].scatter(adult['capital-loss'],adult['income'])
ax[2][2].scatter(adult['hours-per-week'],adult['income'])
sns.countplot(y = 'native-country', hue = 'income', data = cat_features,ax=ax[2][3])
```

```python
adult['workclass']=adult['workclass'].astype('category')
adult['marital-status']=adult['marital-status'].astype('category')
adult['occupation']=adult['occupation'].astype('category')
adult['relationship']=adult['relationship'].astype('category')
adult['race']=adult['race'].astype('category')
adult['gender']=adult['gender'].astype('category')
adult['native-country']=adult['relationship'].astype('category')
adult['income']=adult['income'].astype('category')
```
